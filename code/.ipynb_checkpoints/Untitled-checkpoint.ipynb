```python
import pandas as pd
import numpy as np
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.metrics.pairwise import cosine_similarity
```

```python
master = pd.read_csv('../data/master_prod.csv')
master.head()
```

```python
master.drop(columns = ['Unnamed: 0','index'],inplace = True)
master.head(10)
```

```python
review = pd.read_csv('../data/year_2018.csv')
review.drop(columns = ['Unnamed: 0'],inplace = True)
review.head()
```

```python
df = pd.merge(master, review, on='asin')
df.shape
```

```python
df.head(10)
```

```python
df['custom_rating'].dtype
```

```python
df = df[['asin','product_name','overall','reviewerID']].copy()
df.head(10)
```

```python
df.rename(columns = {'overall':'custom_rating'},inplace = True)
df.head(10)
```

```python
pivot = pd.pivot_table(df, index='asin', columns='reviewerID', values='custom_rating')
pivot.head()
```

```python
sparse_pivot = sparse.csr_matrix(pivot.fillna(0))
print(sparse_pivot)
```

```python
cosine = cosine_similarity(sparse_pivot)
```

```python
cosine
```

```python
recommender_user_asin = pairwise_distances(sparse_pivot, metric='cosine')
```

```python
recommender_user_asin.round(8)
```

```python
recommender_user_asin_df = pd.DataFrame(recommender_user_asin, columns=pivot.index, index=pivot.index)
recommender_user_asin_df.head()
```

```python
recommender_user_asin_df.round(8)
```

```python
search = 'B018ETVAZW'

for num in master.loc[master['asin'].str.contains(search), 'asin']:
    print(num)
    #print('Average rating', pivot.loc[title, :].mean())
    #print('Number of ratings', pivot.T[title].count())
    print('')
    print('10 closest products')
    print(round(recommender_user_asin_df[num].sort_values(),12)[100:110])
    print('')
    print('*******************************************************************************************')
    print('')
```

```python
pivot_name = pd.pivot_table(df, index='product_name', columns='reviewerID', values='custom_rating')
pivot_name.head()
```

```python
sparse_pivot_name = sparse.csr_matrix(pivot_name.fillna(0))
print(sparse_pivot_name)
```

```python
recommender_user_name = pairwise_distances(sparse_pivot_name, metric='cosine')
recommender_user_name
```

```python
recommender_user_name_df = pd.DataFrame(recommender_user_name, columns=pivot_name.index, index=pivot_name.index)
recommender_user_name_df.head()
```
