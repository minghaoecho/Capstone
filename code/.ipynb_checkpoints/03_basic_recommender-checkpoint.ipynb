## combine all files together

```python
file_paths = []

for file in os.listdir('../data'):
    if 'product' in file:
        file_paths.append('../data/'+file)
file_paths
```

```python
master_product = {
    'Unnamed: 0': [],
    'asin': [],
    'product_name': [],
    'product_brand': [],
    'product_rating': [],
    'product_rate_count': [],
    'category': [],
    'description': [],
}

master_product = pd.DataFrame(master_product)
```

```python
print(master_product.shape)

master_product = master_product.sort_values(
    ['asin'], ascending=False).drop_duplicates(
        subset = ['asin'], keep ='first')

print(master_product.shape)
```

```python
tfidf = TfidfVectorizer(stop_words='english',min_df=2,max_df=.90,ngram_range=(1,2))
```

```python
term_matrix = tfidf.fit_transform(desc_cate['description'])
```

```python
term_matrix
```

```python
type(term_matrix)
```

```python
term_df = pd.DataFrame(term_matrix.toarray(), columns=tfidf.get_feature_names())

term_df.head()
```

```python
term_df.shape
```

```python
kmeans = KMeans(n_clusters = 500,random_state = 42,n_jobs = 3).fit(term_matrix)
```

```python
silhouette_score(term_matrix, kmeans.labels_)
```

```python
kmeans = KMeans(n_clusters = 2000,random_state = 42).fit(term_matrix)
```

```python
silhouette_score(term_matrix, kmeans.labels_)
```

```python
scores = []
for k in range(2000,4000):
    cl = KMeans(n_clusters=k)
    cl.fit(term_matrix)
    inertia = cl.inertia_
    sil = silhouette_score(term_matrix, cl.labels_)
    scores.append([k, inertia, sil])
    
score_df = pd.DataFrame(scores)
score_df.columns = ['k', 'inertia', 'silhouette']
```
